```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from random import choice, randint

import os

XPATH = By.XPATH

save_path = ''
how_many2save = 1000

custom_user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36"
```

```python
chrome_options = Options()
# chrome_options.add_argument('--headless')

chrome_options.add_argument(f"--user-agent={custom_user_agent}")

driver = webdriver.Chrome(options=chrome_options)
```

```python
def parse(url : str):
    driver.get(url)

    try:
        image = driver.find_element(XPATH, '//div[@class="under-image"]/img')
        screenshot_id = url.split('/')[-1]
        # print(image.get_attribute('src').split('/')[-1])

        if len(image.get_attribute('src').split('/')[-1]) == len(screenshot_id):
            with open(f'Screenshots/screenshot_{screenshot_id}.png', 'wb')  as saving_screenshot:
                saving_screenshot.write(image.screenshot_as_png)

            return(True)

        else:
            print(f'Skipped {url}')
            return(False)

    except Exception as exc:
        # print(str(exc))
        return(False)
```

```python
alphabet = '0123456789abcdefghijklmnopqrstuvxywz'


ids_visited = {}
saved_screenshots = 0

while saved_screenshots != how_many2save:
    id = ''

    for i in range(11):
        id += choice(alphabet).upper() if randint(0, 1) else choice(alphabet)
    
    

    if id not in ids_visited:
        result = parse(f'https://prnt.sc/{id}')

        if result:
            print(f'Saved {id}')
            saved_screenshots += 1
        
        ids_visited.update({id : result})
        
```
